```python
print("═" * 80)
print("🔀 CHAPTER 3: YOUR THREE PATHS")
print("═" * 80)

# Tax calculation function
def calculate_tax_mfj(taxable_income):
    """Calculate federal tax for Married Filing Jointly (2024 brackets)"""
    brackets = [
        (23200, 0.10), (94300, 0.12), (201050, 0.22),
        (383900, 0.24), (487450, 0.32), (731200, 0.35), (float('inf'), 0.37)
    ]
    tax = 0
    prev = 0
    for limit, rate in brackets:
        if taxable_income <= prev:
            break
        taxable_in_bracket = min(taxable_income, limit) - prev
        tax += taxable_in_bracket * rate
        prev = limit
    return tax

# Project each path over 25 years
years = 25

# PATH A: Do Nothing
path_a_ira = TOTAL_PRETAX
path_a_roth = TOTAL_ROTH
path_a_taxable = JOINT_TAXABLE_ACCOUNTS

for yr in range(years):
    # Grow accounts
    path_a_ira *= (1 + IRA_RETURN)
    path_a_roth *= (1 + ROTH_RETURN)
    path_a_taxable *= (1 + TAXABLE_RETURN)
    
    # Withdraw for income (simplified)
    income_need = ANNUAL_INCOME_NEED * (1 + INFLATION_RATE) ** yr
    ss1 = SPOUSE1_SS_ANNUAL if yr >= years_to_rajesh_ss else 0
    ss2 = SPOUSE2_SS_ANNUAL if yr >= years_to_terri_ss else 0
    from_savings = max(0, income_need - ss1 - ss2)
    path_a_ira -= from_savings * 0.7  # Most from IRA
    path_a_taxable -= from_savings * 0.3 * 0.22  # Taxes

path_a_after_tax = path_a_ira * 0.75 + path_a_roth + path_a_taxable * 0.92

# PATH B: Smart Conversions ($500K over 5 years)
conversion_total_b = 500_000
conversion_tax_b = conversion_total_b * 0.24

path_b_ira = TOTAL_PRETAX - conversion_total_b
path_b_roth = TOTAL_ROTH + conversion_total_b
path_b_taxable = JOINT_TAXABLE_ACCOUNTS - conversion_tax_b

for yr in range(years):
    path_b_ira *= (1 + IRA_RETURN)
    path_b_roth *= (1 + ROTH_RETURN)
    path_b_taxable *= (1 + TAXABLE_RETURN)
    
    income_need = ANNUAL_INCOME_NEED * (1 + INFLATION_RATE) ** yr
    ss1 = SPOUSE1_SS_ANNUAL if yr >= years_to_rajesh_ss else 0
    ss2 = SPOUSE2_SS_ANNUAL if yr >= years_to_terri_ss else 0
    from_savings = max(0, income_need - ss1 - ss2)
    path_b_ira -= from_savings * 0.5
    path_b_roth -= from_savings * 0.2
    path_b_taxable -= from_savings * 0.3 * 0.15

path_b_after_tax = path_b_ira * 0.75 + path_b_roth + path_b_taxable * 0.92

# PATH C: Aggressive Conversions ($1.5M over 10 years)
conversion_total_c = 1_500_000
conversion_tax_c = conversion_total_c * 0.26

path_c_ira = TOTAL_PRETAX - conversion_total_c
path_c_roth = TOTAL_ROTH + conversion_total_c
path_c_taxable = max(MINIMUM_CASH_RESERVE, JOINT_TAXABLE_ACCOUNTS - conversion_tax_c)

for yr in range(years):
    path_c_ira *= (1 + IRA_RETURN)
    path_c_roth *= (1 + ROTH_RETURN)
    path_c_taxable *= (1 + TAXABLE_RETURN)
    
    income_need = ANNUAL_INCOME_NEED * (1 + INFLATION_RATE) ** yr
    ss1 = SPOUSE1_SS_ANNUAL if yr >= years_to_rajesh_ss else 0
    ss2 = SPOUSE2_SS_ANNUAL if yr >= years_to_terri_ss else 0
    from_savings = max(0, income_need - ss1 - ss2)
    path_c_roth -= from_savings * 0.7
    path_c_ira -= from_savings * 0.3

path_c_after_tax = path_c_ira * 0.75 + path_c_roth + path_c_taxable * 0.92

# First RMD calculations
rmd_divisor = 26.5
path_a_first_rmd = (TOTAL_PRETAX * (1 + IRA_RETURN) ** 12) / rmd_divisor
path_b_first_rmd = ((TOTAL_PRETAX - conversion_total_b) * (1 + IRA_RETURN) ** 12) / rmd_divisor
path_c_first_rmd = ((TOTAL_PRETAX - conversion_total_c) * (1 + IRA_RETURN) ** 12) / rmd_divisor

print(f"""
┌─────────────────────────────────────────────────────────────────────────────┐
│                         THE THREE PATHS COMPARED                            │
└─────────────────────────────────────────────────────────────────────────────┘

╔═══════════════════════╦═══════════════════╦═══════════════════╦═══════════════════╗
║                       ║   PATH A:         ║   PATH B:         ║   PATH C:         ║
║       METRIC          ║   Do Nothing      ║   Smart Convert   ║   Aggressive      ║
╠═══════════════════════╬═══════════════════╬═══════════════════╬═══════════════════╣
║ Roth Conversions      ║ ${0:>14,.0f}  ║ ${conversion_total_b:>14,.0f}  ║ ${conversion_total_c:>14,.0f}  ║
║ Conversion Taxes      ║ ${0:>14,.0f}  ║ ${conversion_tax_b:>14,.0f}  ║ ${conversion_tax_c:>14,.0f}  ║
╠═══════════════════════╬═══════════════════╬═══════════════════╬═══════════════════╣
║ IRA at Year 25        ║ ${path_a_ira:>14,.0f}  ║ ${path_b_ira:>14,.0f}  ║ ${path_c_ira:>14,.0f}  ║
║ Roth at Year 25       ║ ${path_a_roth:>14,.0f}  ║ ${path_b_roth:>14,.0f}  ║ ${path_c_roth:>14,.0f}  ║
║ Taxable at Year 25    ║ ${path_a_taxable:>14,.0f}  ║ ${path_b_taxable:>14,.0f}  ║ ${path_c_taxable:>14,.0f}  ║
╠═══════════════════════╬═══════════════════╬═══════════════════╬═══════════════════╣
║ AFTER-TAX WEALTH      ║ ${path_a_after_tax:>14,.0f}  ║ ${path_b_after_tax:>14,.0f}  ║ ${path_c_after_tax:>14,.0f}  ║
║ vs Do Nothing         ║ ${0:>14,.0f}  ║ ${path_b_after_tax - path_a_after_tax:>+14,.0f}  ║ ${path_c_after_tax - path_a_after_tax:>+14,.0f}  ║
╠═══════════════════════╬═══════════════════╬═══════════════════╬═══════════════════╣
║ First RMD (Age 73)    ║ ${path_a_first_rmd:>14,.0f}  ║ ${path_b_first_rmd:>14,.0f}  ║ ${path_c_first_rmd:>14,.0f}  ║
╚═══════════════════════╩═══════════════════╩═══════════════════╩═══════════════════╝

  💡 KEY INSIGHT:
     • Path A leaves the most in IRAs, but biggest RMDs and tax burden
     • Path B is the sweet spot - significant RMD reduction, affordable taxes
     • Path C reduces RMDs most, but may strain taxable account
""")
```

```python
# =============================================================================
# 🏠 WHAT IF: HOME PURCHASE REQUIRING $200K DOWN PAYMENT
# =============================================================================

print("═" * 80)
print("🏠 CHAPTER 8: WHAT IF - HOME PURCHASE ($200K DOWN PAYMENT)")
print("═" * 80)

# The scenario
HOME_DOWN_PAYMENT = 200_000

print(f"""
┌─────────────────────────────────────────────────────────────────────────────┐
│                         THE HOME PURCHASE SCENARIO                          │
└─────────────────────────────────────────────────────────────────────────────┘

  🏠 SCENARIO: You want to purchase a new home and need $200,000 for a 
     down payment to keep monthly payments stable/affordable.

  💵 WHERE DOES THE $200K COME FROM?

  Your current liquid assets:
  ├── Taxable Account:  ${JOINT_TAXABLE_ACCOUNTS:>12,.0f}
  ├── IRAs (pre-tax):   ${TOTAL_PRETAX:>12,.0f}  ← Withdrawal = income tax!
  └── Roth IRAs:        ${TOTAL_ROTH:>12,.0f}

  ⚠️  THE PROBLEM: 
     If you use taxable for the down payment, you won't have funds
     to pay Roth conversion taxes!
""")

# Option analysis
print("""
┌─────────────────────────────────────────────────────────────────────────────┐
│                    OPTIONS FOR FUNDING THE DOWN PAYMENT                     │
└─────────────────────────────────────────────────────────────────────────────┘
""")

# OPTION 1: Use taxable account
taxable_after_purchase = JOINT_TAXABLE_ACCOUNTS - HOME_DOWN_PAYMENT
print(f"""
  ═══════════════════════════════════════════════════════════════════════════
  OPTION 1: Use Taxable Account for Down Payment
  ═══════════════════════════════════════════════════════════════════════════
  
  • Down payment:           ${HOME_DOWN_PAYMENT:>12,.0f}
  • Taxable remaining:      ${taxable_after_purchase:>12,.0f}
  • Cash reserve needed:    ${MINIMUM_CASH_RESERVE:>12,.0f}
  • Available for taxes:    ${taxable_after_purchase - MINIMUM_CASH_RESERVE:>12,.0f}
  
  ⚠️  IMPACT ON ROTH CONVERSIONS:
""")

if taxable_after_purchase - MINIMUM_CASH_RESERVE > 50000:
    max_conversion_possible = (taxable_after_purchase - MINIMUM_CASH_RESERVE) / 0.24
    print(f"""     • You can still convert ~${max_conversion_possible:,.0f} total
     • That's ~${max_conversion_possible/4:,.0f}/year for 4 years
     • MUCH less than the recommended $500K
""")
else:
    print(f"""     • ❌ NOT ENOUGH for meaningful conversions!
     • Only ${taxable_after_purchase - MINIMUM_CASH_RESERVE:,.0f} available after reserve
     • Conversions would have to wait or be minimal
""")

# OPTION 2: Use IRA withdrawal
ira_withdrawal_for_home = HOME_DOWN_PAYMENT
tax_on_ira_withdrawal = ira_withdrawal_for_home * 0.26  # ~26% effective rate
gross_ira_needed = HOME_DOWN_PAYMENT + tax_on_ira_withdrawal

print(f"""
  ═══════════════════════════════════════════════════════════════════════════
  OPTION 2: Withdraw from IRA for Down Payment
  ═══════════════════════════════════════════════════════════════════════════
  
  To get ${HOME_DOWN_PAYMENT:,} cash, you need to withdraw more (for taxes):
  
  • Gross IRA withdrawal:   ${gross_ira_needed:>12,.0f}
  • Tax on withdrawal:      ${tax_on_ira_withdrawal:>12,.0f}
  • Net for down payment:   ${HOME_DOWN_PAYMENT:>12,.0f}
  
  • Taxable preserved:      ${JOINT_TAXABLE_ACCOUNTS:>12,.0f}
  • Available for conv tax: ${JOINT_TAXABLE_ACCOUNTS - MINIMUM_CASH_RESERVE:>12,.0f}
  
  ✅ BENEFIT: Preserves taxable for Roth conversion taxes
  ❌ COST: Already "using up" your low tax brackets this year
           Conversion capacity reduced in year of withdrawal
""")

# OPTION 3: Split approach
split_from_taxable = 100_000
split_from_ira = 100_000
tax_on_split_ira = split_from_ira * 0.24
gross_ira_for_split = split_from_ira + tax_on_split_ira

print(f"""
  ═══════════════════════════════════════════════════════════════════════════
  OPTION 3: Split Between Taxable and IRA (Recommended)
  ═══════════════════════════════════════════════════════════════════════════
  
  Take ${split_from_taxable:,} from taxable + ${split_from_ira:,} from IRA:
  
  From Taxable:
  • Use:                    ${split_from_taxable:>12,.0f}
  • Remaining:              ${JOINT_TAXABLE_ACCOUNTS - split_from_taxable:>12,.0f}
  
  From IRA (grossed up for tax):
  • Gross withdrawal:       ${gross_ira_for_split:>12,.0f}
  • Tax:                    ${tax_on_split_ira:>12,.0f}
  • Net for home:           ${split_from_ira:>12,.0f}
  
  TOTAL for down payment:   ${split_from_taxable + split_from_ira:>12,.0f} ✓
  
  Taxable after purchase:   ${JOINT_TAXABLE_ACCOUNTS - split_from_taxable:>12,.0f}
  Available for conv taxes: ${JOINT_TAXABLE_ACCOUNTS - split_from_taxable - MINIMUM_CASH_RESERVE:>12,.0f}
  
  ✅ This preserves ~${JOINT_TAXABLE_ACCOUNTS - split_from_taxable - MINIMUM_CASH_RESERVE:,} for conversion taxes
""")

# Impact comparison
print("""
┌─────────────────────────────────────────────────────────────────────────────┐
│                    IMPACT ON YOUR RETIREMENT PLAN                           │
└─────────────────────────────────────────────────────────────────────────────┘
""")

# Calculate scenarios
# SCENARIO A: No home purchase (original plan)
scenario_a_conversions = 500_000
scenario_a_conv_tax = scenario_a_conversions * 0.24

# SCENARIO B: Home purchase using taxable
scenario_b_taxable_remaining = taxable_after_purchase - MINIMUM_CASH_RESERVE
scenario_b_conversions = max(0, scenario_b_taxable_remaining / 0.24)
scenario_b_conv_tax = scenario_b_conversions * 0.24

# SCENARIO C: Home purchase with split approach
scenario_c_taxable_remaining = JOINT_TAXABLE_ACCOUNTS - split_from_taxable - MINIMUM_CASH_RESERVE
scenario_c_conversions = scenario_c_taxable_remaining / 0.24  # What can be funded
scenario_c_conversions = min(400_000, scenario_c_conversions)  # Cap at reasonable amount
scenario_c_conv_tax = scenario_c_conversions * 0.24

# Project 25-year outcomes
def project_wealth(initial_ira, initial_roth, initial_taxable, conversions):
    ira = initial_ira - conversions
    roth = initial_roth + conversions
    taxable = initial_taxable - (conversions * 0.24)
    
    for yr in range(25):
        ira *= (1 + IRA_RETURN)
        roth *= (1 + ROTH_RETURN)
        taxable *= (1 + TAXABLE_RETURN)
        
        # Income draws (simplified)
        income_need = ANNUAL_INCOME_NEED * (1 + INFLATION_RATE) ** yr
        ss1 = SPOUSE1_SS_ANNUAL if yr >= years_to_rajesh_ss else 0
        ss2 = SPOUSE2_SS_ANNUAL if yr >= years_to_terri_ss else 0
        from_savings = max(0, income_need - ss1 - ss2)
        ira -= from_savings * 0.5
        taxable -= from_savings * 0.2
    
    after_tax = ira * 0.75 + roth + taxable * 0.92
    return ira, roth, taxable, after_tax

# Scenario A: Original plan (no home)
a_ira, a_roth, a_tax, a_total = project_wealth(
    TOTAL_PRETAX, TOTAL_ROTH, JOINT_TAXABLE_ACCOUNTS, scenario_a_conversions)

# Scenario B: Home, use taxable
b_ira, b_roth, b_tax, b_total = project_wealth(
    TOTAL_PRETAX, TOTAL_ROTH, taxable_after_purchase, scenario_b_conversions)

# Scenario C: Home, split approach (IRA withdrawal reduces IRA base)
c_ira, c_roth, c_tax, c_total = project_wealth(
    TOTAL_PRETAX - gross_ira_for_split, TOTAL_ROTH, 
    JOINT_TAXABLE_ACCOUNTS - split_from_taxable, scenario_c_conversions)

print(f"""
╔════════════════════════════╦════════════════╦════════════════╦════════════════╗
║                            ║  NO HOME       ║  HOME (Taxable)║  HOME (Split)  ║
║         METRIC             ║  (Original)    ║  for Down Pmt  ║  Tax+IRA       ║
╠════════════════════════════╬════════════════╬════════════════╬════════════════╣
║ Down Payment Source        ║     N/A        ║  Taxable       ║  $100K each    ║
║ Taxable Remaining          ║ ${JOINT_TAXABLE_ACCOUNTS:>12,}  ║ ${taxable_after_purchase:>12,}  ║ ${JOINT_TAXABLE_ACCOUNTS - split_from_taxable:>12,}  ║
║ Roth Conversions Possible  ║ ${scenario_a_conversions:>12,}  ║ ${scenario_b_conversions:>12,.0f}  ║ ${scenario_c_conversions:>12,.0f}  ║
╠════════════════════════════╬════════════════╬════════════════╬════════════════╣
║ 25-Year After-Tax Wealth   ║ ${a_total:>12,.0f}  ║ ${b_total:>12,.0f}  ║ ${c_total:>12,.0f}  ║
║ vs Original Plan           ║ ${0:>12,}  ║ ${b_total - a_total:>+12,.0f}  ║ ${c_total - a_total:>+12,.0f}  ║
╚════════════════════════════╩════════════════╩════════════════╩════════════════╝
""")

# Legacy impact
kids_tax = 0.28
a_legacy = a_ira * (1-kids_tax) + a_roth + a_tax * 0.95
b_legacy = b_ira * (1-kids_tax) + b_roth + b_tax * 0.95
c_legacy = c_ira * (1-kids_tax) + c_roth + c_tax * 0.95

print(f"""
  📊 IMPACT ON LEGACY (what kids receive after their taxes):
  
     • No Home (Original):   ${a_legacy:>12,.0f}
     • Home (Taxable):       ${b_legacy:>12,.0f}  ({b_legacy - a_legacy:+,.0f})
     • Home (Split):         ${c_legacy:>12,.0f}  ({c_legacy - a_legacy:+,.0f})
""")

# Recommendation
print("""
┌─────────────────────────────────────────────────────────────────────────────┐
│                           RECOMMENDATION                                    │
└─────────────────────────────────────────────────────────────────────────────┘
""")

print(f"""
  🏠 IF YOU BUY THE HOME:

  ✅ USE THE SPLIT APPROACH:
     • Take $100,000 from taxable account
     • Withdraw $124,000 from IRA ($100K + $24K tax)
     • Preserves ${JOINT_TAXABLE_ACCOUNTS - split_from_taxable:,} in taxable for conversion taxes
     • Still allows ~${scenario_c_conversions:,.0f} in Roth conversions over 4 years

  📊 THE COST OF BUYING THE HOME:
     • Reduces 25-year wealth by ~${a_total - c_total:,.0f}
     • Reduces legacy to kids by ~${a_legacy - c_legacy:,.0f}
     • Reduces Roth conversions by ~${scenario_a_conversions - scenario_c_conversions:,.0f}

  🤔 IS IT WORTH IT?
     This is a LIFESTYLE decision, not just financial:
     • If the new home improves your quality of life → priceless
     • The "cost" is about ${(a_total - c_total)/25:,.0f}/year in wealth accumulation
     • You're still in good shape - just with modified conversion targets

  📋 MODIFIED CONVERSION PLAN (with home purchase in Year 1):
  
     Year 1:  $75,000 - $100,000 (reduced due to IRA withdrawal for home)
     Year 2:  $100,000
     Year 3:  $100,000  
     Year 4:  $100,000
     ─────────────────────
     TOTAL:   ~$375,000 - $400,000 (vs $500K without home purchase)

  ⚠️  KEY: Don't skip conversions entirely just because you bought a home!
     Even ${scenario_c_conversions:,.0f} in conversions is much better than $0.
""")

# =============================================================================
# 📅 TIMING ANALYSIS: WHAT IF YOU DELAY THE HOME PURCHASE?
# =============================================================================

print("""
┌─────────────────────────────────────────────────────────────────────────────┐
│              📅 TIMING MATTERS: WHAT IF YOU DELAY THE PURCHASE?             │
└─────────────────────────────────────────────────────────────────────────────┘
""")

def project_with_timing(purchase_year, down_payment, initial_ira, initial_roth, initial_taxable):
    """
    Simulate: Do conversions BEFORE buying, then purchase in specified year.
    purchase_year: 0 = buy immediately, 1 = buy after 1 year, 2 = buy after 2 years
    """
    ira = initial_ira
    roth = initial_roth
    taxable = initial_taxable
    total_conversions = 0
    conversion_schedule = []
    
    # Pre-purchase years: maximize conversions
    for yr in range(purchase_year):
        # Convert as much as possible while keeping cash reserve + down payment protected
        available_for_tax = taxable - MINIMUM_CASH_RESERVE - down_payment
        if available_for_tax > 0:
            max_conversion = min(available_for_tax / 0.24, 150_000)  # Cap annual conversion
            conversion = max(0, max_conversion)
        else:
            conversion = 0
        
        conversion_tax = conversion * 0.24
        ira -= conversion
        roth += conversion
        taxable -= conversion_tax
        total_conversions += conversion
        conversion_schedule.append(conversion)
        
        # Growth
        ira *= (1 + IRA_RETURN)
        roth *= (1 + ROTH_RETURN)
        taxable *= (1 + TAXABLE_RETURN)
    
    # Purchase year: use split approach (100K taxable + 124K IRA)
    taxable -= 100_000  # From taxable
    ira -= 124_000      # Gross from IRA (100K net + 24K tax)
    
    # Purchase year conversion (reduced due to IRA withdrawal)
    available_for_tax = taxable - MINIMUM_CASH_RESERVE
    if available_for_tax > 0:
        conversion = min(available_for_tax / 0.24, 75_000)  # Reduced year
    else:
        conversion = 0
    
    ira -= conversion
    roth += conversion
    taxable -= conversion * 0.24
    total_conversions += conversion
    conversion_schedule.append(conversion)
    
    # Growth
    ira *= (1 + IRA_RETURN)
    roth *= (1 + ROTH_RETURN)
    taxable *= (1 + TAXABLE_RETURN)
    
    # Post-purchase years (up to year 4): continue conversions
    for yr in range(purchase_year + 1, 5):
        available_for_tax = taxable - MINIMUM_CASH_RESERVE
        if available_for_tax > 0:
            conversion = min(available_for_tax / 0.24, 125_000)
        else:
            conversion = 0
        
        ira -= conversion
        roth += conversion
        taxable -= conversion * 0.24
        total_conversions += conversion
        conversion_schedule.append(conversion)
        
        # Growth
        ira *= (1 + IRA_RETURN)
        roth *= (1 + ROTH_RETURN)
        taxable *= (1 + TAXABLE_RETURN)
    
    # Years 5-25: just growth and withdrawals
    for yr in range(5, 25):
        ira *= (1 + IRA_RETURN)
        roth *= (1 + ROTH_RETURN)
        taxable *= (1 + TAXABLE_RETURN)
        
        # Simplified withdrawals
        income_need = ANNUAL_INCOME_NEED * (1 + INFLATION_RATE) ** yr
        ss1 = SPOUSE1_SS_ANNUAL if yr >= years_to_rajesh_ss else 0
        ss2 = SPOUSE2_SS_ANNUAL if yr >= years_to_terri_ss else 0
        from_savings = max(0, income_need - ss1 - ss2)
        ira -= from_savings * 0.5
        taxable -= from_savings * 0.2
    
    after_tax = ira * 0.75 + roth + taxable * 0.92
    legacy = ira * (1 - kids_tax) + roth + taxable * 0.95
    
    return {
        'total_conversions': total_conversions,
        'schedule': conversion_schedule,
        'ira': ira,
        'roth': roth,
        'taxable': taxable,
        'after_tax': after_tax,
        'legacy': legacy
    }

# Calculate for different timing scenarios
timing_results = {}
for delay in [0, 1, 2]:
    timing_results[delay] = project_with_timing(
        delay, HOME_DOWN_PAYMENT, TOTAL_PRETAX, TOTAL_ROTH, JOINT_TAXABLE_ACCOUNTS
    )

print(f"""
  🤔 THE BIG QUESTION: Does it matter WHEN you buy the home?
  
     If you can rent or stay put for 1-2 years while doing conversions first,
     you can build more Roth assets before the down payment depletes your
     taxable account.
""")

print(f"""
╔═══════════════════════════════════════════════════════════════════════════════╗
║                    TIMING COMPARISON: WHEN TO BUY THE HOME                    ║
╠═══════════════════════╦═══════════════╦═══════════════╦═══════════════════════╣
║      SCENARIO         ║  BUY NOW      ║  DELAY 1 YEAR ║  DELAY 2 YEARS        ║
║                       ║  (Year 1)     ║  (Year 2)     ║  (Year 3)             ║
╠═══════════════════════╬═══════════════╬═══════════════╬═══════════════════════╣
║ Total Conversions     ║ ${timing_results[0]['total_conversions']:>11,.0f}  ║ ${timing_results[1]['total_conversions']:>11,.0f}  ║ ${timing_results[2]['total_conversions']:>17,.0f}  ║
╠═══════════════════════╬═══════════════╬═══════════════╬═══════════════════════╣
║ Year-by-Year Plan:    ║               ║               ║                       ║""")

# Show conversion schedules
for i in range(5):
    yr_label = f"  Year {i+1}"
    buy0 = timing_results[0]['schedule'][i] if i < len(timing_results[0]['schedule']) else 0
    buy1 = timing_results[1]['schedule'][i] if i < len(timing_results[1]['schedule']) else 0
    buy2 = timing_results[2]['schedule'][i] if i < len(timing_results[2]['schedule']) else 0
    
    note0 = " 🏠" if i == 0 else ""
    note1 = " 🏠" if i == 1 else ""
    note2 = " 🏠" if i == 2 else ""
    
    print(f"║ {yr_label:<21} ║ ${buy0:>9,.0f}{note0:<2} ║ ${buy1:>9,.0f}{note1:<2} ║ ${buy2:>17,.0f}{note2:<2}  ║")

print(f"""╠═══════════════════════╬═══════════════╬═══════════════╬═══════════════════════╣
║ 25-Year Wealth        ║ ${timing_results[0]['after_tax']:>11,.0f}  ║ ${timing_results[1]['after_tax']:>11,.0f}  ║ ${timing_results[2]['after_tax']:>17,.0f}  ║
║ Legacy to Kids        ║ ${timing_results[0]['legacy']:>11,.0f}  ║ ${timing_results[1]['legacy']:>11,.0f}  ║ ${timing_results[2]['legacy']:>17,.0f}  ║
╠═══════════════════════╬═══════════════╬═══════════════╬═══════════════════════╣
║ vs Buy Now            ║ $          0  ║ ${timing_results[1]['after_tax'] - timing_results[0]['after_tax']:>+11,.0f}  ║ ${timing_results[2]['after_tax'] - timing_results[0]['after_tax']:>+17,.0f}  ║
╚═══════════════════════╩═══════════════╩═══════════════╩═══════════════════════╝
""")

# Calculate the benefit of waiting
benefit_1yr = timing_results[1]['after_tax'] - timing_results[0]['after_tax']
benefit_2yr = timing_results[2]['after_tax'] - timing_results[0]['after_tax']
extra_conv_1yr = timing_results[1]['total_conversions'] - timing_results[0]['total_conversions']
extra_conv_2yr = timing_results[2]['total_conversions'] - timing_results[0]['total_conversions']

print(f"""
  📊 THE VALUE OF WAITING:
  
     ┌────────────────────────────────────────────────────────────────────────┐
     │  DELAY 1 YEAR:                                                        │
     │  • Extra Roth conversions:  ${extra_conv_1yr:>+12,.0f}                            │
     │  • Extra 25-year wealth:    ${benefit_1yr:>+12,.0f}                            │
     │  • That's ${benefit_1yr/1:>+,.0f} for waiting 12 months!                     │
     └────────────────────────────────────────────────────────────────────────┘
     
     ┌────────────────────────────────────────────────────────────────────────┐
     │  DELAY 2 YEARS:                                                       │
     │  • Extra Roth conversions:  ${extra_conv_2yr:>+12,.0f}                            │
     │  • Extra 25-year wealth:    ${benefit_2yr:>+12,.0f}                            │
     │  • That's ${benefit_2yr/2:>+,.0f}/year for waiting!                          │
     └────────────────────────────────────────────────────────────────────────┘
""")

# Final timing recommendation
print("""
┌─────────────────────────────────────────────────────────────────────────────┐
│                       🎯 TIMING RECOMMENDATION                              │
└─────────────────────────────────────────────────────────────────────────────┘
""")

if benefit_1yr > 50000:
    print(f"""
  ⭐ BEST STRATEGY: DELAY THE HOME PURCHASE BY 1-2 YEARS IF POSSIBLE
  
  WHY? The first year or two of Roth conversions are the MOST valuable
  because:
  
  1. You have the most taxable funds available to pay conversion taxes
  2. The converted money has the longest time to grow tax-free
  3. Once you use $200K for the down payment, conversion capacity drops
  
  📋 OPTIMAL PLAN (if you can delay 1 year):
  
     Year 1:  Convert $125,000 - $150,000 (BIG year before home purchase)
     Year 2:  Buy home 🏠 + Convert $75,000 (reduced year)
     Year 3:  Convert $100,000
     Year 4:  Convert $100,000
     ─────────────────────────────────────
     TOTAL:   ~${timing_results[1]['total_conversions']:,.0f} in conversions
     
  💰 VALUE: This approach gives you ${benefit_1yr:,.0f} more in 25-year wealth!
""")
else:
    print(f"""
  📊 Delaying provides modest benefit of ${benefit_1yr:,.0f}.
     If the home purchase is urgent, proceed with the split approach.
""")

print(f"""
  ⚠️  HOWEVER - LIFE CONSIDERATIONS:
  
     • Can you rent affordably for 1-2 years?
     • Are home prices rising faster than your conversion gains?
     • Is there an ideal home available NOW that might not be later?
     • What's the stress/hassle cost of delaying?
     
  💡 THE MATH SAYS: Wait 1-2 years if possible
     LIFE SAYS: Sometimes the right home at the right time is worth the cost
""")
```
